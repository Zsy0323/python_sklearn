为方便讲解，我们首先定义一个生成DataFrame的函数：

````python
# 列:[ABC]
# index :[123]
#    A   B   C
# 1  A1....
# 2  B1....
# 3  C1....
def make_df(inds,cols):
    data = {key:[key + str(i) for i in inds] for key in cols}
    df = DataFrame(data=data,index=inds)
    return df
````

````python
make_df([1,2,3],list('ABC'))
````

## 1. 使用pd.concat()级联

pandas使用pd.concat函数，与np.concatenate函数类似，只是多了一些参数：
```
objs
axis=0
join='outer'
join_axes=None
ignore_index=False
keys = [value1,value2...]
```

### 1)  简单级联

和np.concatenate一样，优先增加行数（默认axis=0）

````python
df1 = make_df([1,2,3],list('ABC'))
df2 = make_df([2,3,4],list('CDE'))
display(df1,df2)
````

````python
pd.concat((df1,df2))
````

可以通过设置axis来改变级联方向

````python
pd.concat((df1,df2),axis=1)
````

级连会把该方向上索引相同的元素放在一行（一列），index/columns在级联时可以重复

也可以选择忽略ignore_index，重新索引

````python
pd.concat((df1,df2),ignore_index=True)
````

或者使用多层索引 keys  

concat([x,y],keys=['x','y'])

````python
pd.concat((df1,df2),keys=['x','y'])
````

### 2) 不匹配级联

不匹配指的是级联的维度的索引不一致。例如纵向级联时列索引不一致，横向级联时行索引不一致

````python
df1 = make_df([1,2,3,4],list('ABCD'))
df2 = make_df([2,3,4,5,6],list('CDE'))
display(df1,df2)
````

````python
pd.concat((df1,df2))
````

- 外连接：补NaN（默认模式）

- 内连接：只连接匹配的项

````python
# 只连接匹配的项
pd.concat((df1,df2),join='inner')
````

````python
df1.columns
````

## 2. 使用pd.merge()合并

merge与concat的区别在于，merge需要依据某一共同的行或列来进行合并

使用pd.merge()合并时，会自动根据两者相同column名称的那一列，作为key来进行合并。

注意每一列元素的顺序不要求一致

````python
df1 = DataFrame({'name':['张三','李四','mike'],'id':[1,2,3],'age':[21,25,23]})
df2 = DataFrame({'sex':['男','男','女'],'id':[2,3,4],'group':['a','b','c']})
display(df1,df2)
````

###  1) 一对一合并

````python
pd.merge(df1,df2)
````

### 2) 多对一合并

````python
df1 = DataFrame({'name':['张三','李四','Chales'],'id':[1,2,2],'age':[22,21,25]})
df2 = DataFrame({'sex':['男','男','女'],'id':[2,3,4],'group':['a','b','c']})
display(df1,df2)
````

````python
pd.merge(df1,df2)
````

### 3) 多对多合并

````python
df1 = DataFrame({'name':['张三','李四','张三'],'salary':[10000,12000,20000],'age':[22,21,25]})
df2 = DataFrame({'sex':['男','男','女'],'name':['张三','张三','凡凡'],'group':['a','b','c']})
display(df1,df2)
````

````python
pd.merge(df1,df2)
````

### 4) key的规范化

- 使用on=显式指定哪一列为key,当有多个key相同时使用

````python
df1 = DataFrame({'name':['张三','李四','张三'],'salary':[10000,12000,20000],'age':[22,21,25]})
df2 = DataFrame({'age':[21,18,29],'name':['张三','张三','凡凡'],'group':['sale','search','service']})
display(df1,df2)
````

````python
pd.merge(df1,df2,on='name',suffixes=['_1','_2'])
````

- 使用left_on和right_on指定左右两边的列作为key，当左右两边的key都不相等时使用

````python
df1 = DataFrame({'name':['张三','李四','张三'],'salary':[10000,12000,20000],'age':[22,21,25]})
df2 = DataFrame({'年龄':[21,18,29],'名字':['张三','张三','凡凡'],'group':['sale','search','service']})
display(df1,df2)
````

````python
pd.merge(df1,df2,left_on='name',right_on='名字')
````

#### 当左边的列和右边的index相同的时候,使用right_index = True

````python
df1 = DataFrame({'name':['张三','李四','张三'],'salary':[10000,12000,20000],'age':[22,21,25]})
df2 = DataFrame({'年龄':[21,18,29],'名字':['张三','张三','凡凡'],'group':['sale','search','service']
                },index = [22,21,25])
display(df1,df2)
````

````python
pd.merge(df1,df2,left_on='age',right_index=True)
````

### 5) 内合并与外合并

- 内合并：只保留两者都有的key（默认模式）

````python
df1 = DataFrame({'name':['张三','李四','张三'],'salary':[10000,12000,20000],'age':[22,21,25]})
df2 = DataFrame({'age':[21,18,29],'名字':['张三','张三','凡凡'],'group':['sale','search','service']
                })
display(df1,df2)
````

- 外合并 how='outer'：补NaN

````python
pd.merge(df1,df2,how='inner')
````

- 左合并、右合并：how='left'，how='right'，

````python
pd.merge(df1,df2,how='left')
````

````python
pd.merge(df1,df2,how='right')
````

# 3. 案例分析：美国各州人口数据分析

## 首先导入文件，并查看数据样本

````python
s_abb = pd.read_csv('state-abbrevs.csv')
s_abb.head()
````

````python
s_pop = pd.read_csv('state-population.csv')
s_pop.head()
````

````python
s_ares = pd.read_csv('state-areas.csv')
s_ares.head()
````

合并pop与abbrevs两个DataFrame，分别依据state/region列和abbreviation列来合并。

为了保留所有信息，使用外合并。

````python
s_m = pd.merge(s_pop,s_abb,how='outer',left_on='state/region',right_on='abbreviation')
s_m
````

去除abbreviation的那一列（axis=1）

````python
s_d = s_m.drop(['abbreviation'],axis=1)
s_d
````

查看存在缺失数据的列。

使用.isnull().any()，只有某一列存在一个缺失数据，就会显示True。

````python
s_d.isnull().any()
````

查看缺失数据

````python
s_d[s_d.isnull().any(axis=1)]
````

根据数据是否缺失情况显示数据，如果缺失为True，那么显示

````python
s_d.isnull()
````

找到有哪些state/region使得state的值为NaN，使用unique()查看非重复值

````python
cause = s_d['state'].isnull()
s_d['state/region'][cause].unique()
````

为找到的这些state/region的state项补上正确的值，从而去除掉state这一列的所有NaN！

记住这样清除缺失数据NaN的方法！

````python
cause = s_d['state/region'] == 'PR'
s_d['state'][cause] = 'Puerto Rico'
s_d
````

````python
cause = s_d['state/region'] == 'USA'
s_d['state'][cause] = 'United State'
s_d
````

````python
s_d.isnull().any()
````

合并各州面积数据areas，使用左合并。

思考一下为什么使用外合并？



````python
s_p = pd.merge(s_d,s_ares,how='outer')
s_p
````

继续寻找存在缺失数据的列

````python
s_p.isnull()
````

我们会发现area(sq.mi)这一列有缺失数据，为了找出是哪一行，我们需要找出是哪个state没有数据

````python
s_p['state/region'][s_p['area (sq. mi)'].isnull()].unique()
````

去除含有缺失数据的行

````python
s_d2 = s_p.dropna()
s_d2
````

查看数据是否缺失

````python
s_d2.isnull()
````

找出2010年的全民人口数据,df.query(查询语句)

````python
s_p2 = s_d2.query('ages == "total" & year == 2010')
s_p2
````

对查询结果进行处理，以state列作为新的行索引:set_index

````python
s_p3 = s_p2.set_index('state')
s_p3
````

计算人口密度。注意是Series/Series，其结果还是一个Series。

````python
s_m2 = s_p3['population']/s_p3['area (sq. mi)']
s_m2
````

排序，并找出人口密度最高的五个州sort_values()

````python
s_m3 = s_m2.sort_values()
s_m3.tail()
````

找出人口密度最低的五个州

````python
s_m3.head()
````

要点总结：
- 统一用loc()索引
- 善于使用.isnull().any()找到存在NaN的列
- 善于使用.unique()确定该列中哪些key是我们需要的
- 一般使用外合并、左合并，目的只有一个：宁愿该列是NaN也不要丢弃其他列的信息

## 回顾：Series/DataFrame运算与ndarray运算的区别

- Series与DataFrame没有广播，如果对应index没有值，则记为NaN；或者使用add的fill_value来补缺失值
- ndarray有广播，通过重复已有值来计算
